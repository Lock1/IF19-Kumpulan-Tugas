```python
# Tanggal   : 11 November 2019
# Deskripsi : panda makan diagram

import pandas
import matplotlib.pyplot as plot

# Baca file
HAHA = pandas.read_csv("stei-mno.csv")

# Plot histogram
HAHA[["nilai_mat"]].plot(kind="hist",bins=[0,20,40,60,80,100],rwidth=0.8)
```

```python
# Plot bar diagram
HAHA["fakultas"].value_counts().plot(kind="barh",x=HAHA.keys(),y="")
```

```python
# Plot pie chart
HAHA["fakultas"].value_counts().plot(kind="pie").set_ylabel("Fakultas              ")
```

Untuk diagram, menurut saya lebih baik menggunakan diagram batang untuk menunjukan banyaknya mahasiswa per fakultas, karena lebih jelas ketika dibandingkan mahasiswa per fakultasnya.

```python
# Plot scatter
xax = HAHA["nilai_kim"]
yax = HAHA["nilai_fis"]
plot.scatter(xax,yax)
```
