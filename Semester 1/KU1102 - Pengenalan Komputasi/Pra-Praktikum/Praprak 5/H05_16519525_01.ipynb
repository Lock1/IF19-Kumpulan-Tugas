```python
# Tanggal   : 11 November 2019
# Deskripsi : panda makan bambu


import pandas
# Baca file
readit = pandas.read_csv("stei-mno.csv")

# Cari ukuran excel
print("Database memiliki ukuran")
print("Baris :",readit.shape[0])
print("Kolom :",readit.shape[1])
print("Sehingga terdapat total {} data.\n".format(readit.shape[0]*readit.shape[1]))

# Print 10 data pertama
print("10 Data pertama adalah")
print(readit[:10])

# Print data ke-50 hingga 60
print("\nData ke-50 sampai data ke-60 adalah")
print(readit[50:60])

# Cari banyak mahasiswa
print("\nBanyak mahasiswa setiap fakultas")
ky = readit["fakultas"].value_counts().keys()
data = readit["fakultas"].value_counts()
for i in range(len(ky)):
    print("{:5} {:3} ".format(ky[i],data[i]))

# Digunakan corr untuk mencari korelasi
nf = readit["nilai_fis"]
nk = readit["nilai_kim"]
cor = readit["nilai_kim"].corr(readit["nilai_fis"])

if cor > 0.75:
    print("\nNilai fisika berbanding lurus dengan nilai kimia")
elif cor < -0.75:
    print("\nNilai fisika berbanding terbalik dengan nilai kimia")
else:
    print("\nNilai fisika tidak berhubungan dengan kimia")
    
print("Dapat dilihat pada grafik scatter plot pada pertanyaan 3")
```
